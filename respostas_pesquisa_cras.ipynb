```python
import matplotlib.pyplot as plt
import numpy as np

# Dados das perguntas
perguntas = [
    "Melhoria no transporte",
    "Avaliação contínua",
    "Site de agendamento eficaz",
    "Problemas com transporte",
    "Horários convenientes",
    "Funcionários explicam bem",
    "Satisfação com tempo de espera",
    "Equipe acessível",
    "Sentimento de acolhimento"
]

# Porcentagens de respostas "Sim" e "Não"
sim = np.array([3, 0, 98, 5, 0, 92, 37, 38, 43])
nao = np.array([97, 100, 2, 95, 100, 8, 63, 62, 57])

# Posição das barras no eixo x
x = np.arange(len(perguntas))

# Largura das barras
largura = 0.4

# Criando o gráfico
fig, ax = plt.subplots(figsize=(14, 8))
barras1 = ax.bar(x - largura/2, sim, width=largura, label='Sim', color='#4a90e2')
barras2 = ax.bar(x + largura/2, nao, width=largura, label='Não', color='#e94e77')

# Configurando o gráfico
ax.set_ylabel('Porcentagem (%)')
ax.set_title('Respostas da Pesquisa sobre o Serviço de Transporte do CRAS')
ax.set_xticks(x)
ax.set_xticklabels(perguntas, rotation=45, ha='right')
ax.legend()

# Exibindo as porcentagens acima das barras
for barras in [barras1, barras2]:
    for barra in barras:
        altura = barra.get_height()
        ax.annotate(f'{altura:.0f}%',
                    xy=(barra.get_x() + barra.get_width() / 2, altura),
                    xytext=(0, 3),  # deslocamento em relação ao topo
                    textcoords="offset points",
                    ha='center', va='bottom')

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
